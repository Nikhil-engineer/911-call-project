# LGBM

```python
test_x=pd.read_csv("/content/New_X_test.csv")
train_x=pd.read_csv("/content/New_X_train.csv")
test_y=pd.read_csv("/content/New_y_test.csv")
train_y=pd.read_csv("/content/New_y_train.csv")
```

```python
test_x.drop("Unnamed: 0",axis=1,inplace=True)
train_x.drop("Unnamed: 0",axis=1,inplace=True)
```

```python
train_y=train_y["Label"]
test_y=test_y["Label"]
```

```python
train_y
```

```python
# test_y=np.array(test_y)
# train_y=np.array(train_y)
```

```python
train_y
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder 
from pprint import pprint
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.metrics import accuracy_score, classification_report, precision_score, confusion_matrix
from sklearn.model_selection import train_test_split
  
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import normalize
from sklearn.decomposition import PCA
```

```python
# LGBM model on simple dataset

from lightgbm import LGBMClassifier
lgbm=LGBMClassifier()
# Xlgbm_train, Xlgbm_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0,stratify=Y)  
Xlgbm_train=train_x
Xlgbm_test=test_x
y_train=train_y
y_test=test_y
lgbm.fit(Xlgbm_train,y_train)
ylgbm_pred=lgbm.predict(Xlgbm_test)

accuracy_lgbm=accuracy_score(y_test,ylgbm_pred)
print("The accuracy of LGBM model ",accuracy_lgbm)
print()
print(classification_report(y_test,ylgbm_pred ))
print()
cm_lgbm=confusion_matrix(y_test,ylgbm_pred)
print(cm_lgbm)
print()
plt.figure(figsize=(5,5))
sns.heatmap(data=cm_lgbm,linewidths=.5, annot=True,square = True,  cmap = 'Blues')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
all_sample_title = 'Accuracy Score: {0}'.format(lgbm.score(Xlgbm_test, y_test))
plt.title(all_sample_title, size = 15)
```

```python
pd.crosstab(test_y, ylgbm_pred, rownames = ['Actual'], colnames =['Predicted'], margins = True)
```
